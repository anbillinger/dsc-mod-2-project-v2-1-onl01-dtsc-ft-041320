```python
import pandas as pd
import numpy as np
df = pd.read_csv('kc_house_data_cleaned.csv')
df['price_per_sqft']=df['price']/df['sqft_living']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
sns.scatterplot(df.long,df.lat,hue=np.log(df.price_per_sqft))
labels = [12.0,12.8,13.6,14.4,15.2]
labels2 = [round(np.e**x) for x in labels]
#plt.legend(df['labels'])
```
